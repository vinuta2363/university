k-means

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Mall_Customers (1).csv")
print(df)
x=df.iloc[:,[3,4]].values
from sklearn.cluster import KMeans
wcss_list=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42,n_init=10)
  kmeans.fit(x)
  wcss_list.append(kmeans.inertia_)
plt.plot(range(1,11),wcss_list)
plt.title("elbow method")
plt.xlabel("no of clusters")
plt.ylabel("wcss_list")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42,n_init=10)
y_predict=kmeans.fit_predict(x)
plt.scatter(x[y_predict==0, 0], x[y_predict==0, 1], s=100, c='blue', label='cluster1')
plt.scatter(x[y_predict==1, 0], x[y_predict==1, 1], s=100, c='green', label='cluster2')
plt.scatter(x[y_predict==2, 0], x[y_predict==2, 1], s=100, c='pink', label='cluster3')
plt.scatter(x[y_predict==3, 0], x[y_predict==3, 1], s=100, c='red', label='cluster4')
plt.scatter(x[y_predict==4, 0], x[y_predict==4, 1], s=100, c='orange', label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c="yellow",label="centroid")
plt.title("no of clusters")
plt.xlabel("total income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```
