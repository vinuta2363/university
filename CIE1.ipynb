```python
import pandas as pd
import matplotlib.pyplot as plt
mtcars=pd.read_csv('/mtcars.csv')
auto_counts=(mtcars['am']==0).sum()
manual_counts=(mtcars['am']==1).sum()
labels=['Automatic','Manual']
counts=[auto_counts,manual_counts]
plt.bar(labels,counts,color=['red','green'])
plt.title('frequency distribution')
plt.ylabel('frequency')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
mtcars=pd.read_csv('/mtcars.csv')
plt.scatter(mtcars['wt'],mtcars['mpg'],color='purple',alpha=0.6)
plt.title('relationship between wt amd mpg')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
mtcars=pd.read_csv('/mtcars.csv')
auto=(mtcars['am']==0).sum()
manual=(mtcars['am']==1).sum()
plt.bar(['Automatic','Manual'],[auto,manual],color=['purple','red'])
plt.title('common in transmission')
plt.ylabel('transmission')
plt.grid(axis='y')
plt.show()
```
